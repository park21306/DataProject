แบ่งค่าออกแบ็น 2 ส่วนเพื่อแยกค่าอุณหภูมิ เพื่อง่ายต่อการวิเคราะห์

```python
Jan = c['Jan'].str.split("\n",n=1,expand=True)
Feb = c['Feb'].str.split("\n",n=1,expand=True)
Mar = c['Mar'].str.split("\n",n=1,expand=True)
Apr = c['Apr'].str.split("\n",n=1,expand=True)
May = c['May'].str.split("\n",n=1,expand=True)
Jun = c['Jun'].str.split("\n",n=1,expand=True)
Jul = c['Jul'].str.split("\n",n=1,expand=True)
Aug = c['Aug'].str.split("\n",n=1,expand=True)
Sep = c['Sep'].str.split("\n",n=1,expand=True)
Oct = c['Oct'].str.split("\n",n=1,expand=True)
Nov = c['Nov'].str.split("\n",n=1,expand=True)
Dec = c['Dec'].str.split("\n",n=1,expand=True)
Year = c['Year'].str.split("\n",n=1,expand=True)
```

นำค่าที่แบ่งออกไปไว้ใน Column ใหม่ 

```python
data["Jan_C"]= Jan[0]
data["Jan_F"]= Jan[1]
data["Feb_C"]= Feb[0]
data["Feb_F"]= Feb[1]
data["Mar_C"]= Mar[0]
data["Mar_F"]= Mar[1]
data["Apr_C"]= Apr[0]
data["Apr_F"]= Apr[1]
data["May_C"]= May[0]
data["May_F"]= May[1]
data["Jun_C"]= Jun[0]
data["Jun_F"]= Jun[1]
data["Jul_C"]= Jul[0]
data["Jul_F"]= Jul[1]
data["Aug_C"]= Aug[0]
data["Aug_F"]= Aug[1]
data["Sep_C"]= Sep[0]
data["Sep_F"]= Sep[1]
data["Oct_C"]= Oct[0]
data["Oct_F"]= Oct[1]
data["Nov_C"]= Nov[0]
data["Nov_F"]= Nov[1]
data["Dec_C"]= Dec[0]
data["Dec_F"]= Dec[1]
data["Year_C"]= Year[0]
data["Year_F"]= Year[1]
```

ลบ Column ที่มไ่ต้องการออกเนื่องจาก ได้ทำการแบ่งค่า อุณหภูมิออกเป็น องศาเซลเซียส และ ฟาเรนไฮต์
